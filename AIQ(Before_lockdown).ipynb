```python
#importing library

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
#collecting data 

df = pd.read_csv(r"C:\Users\MANISH SHARMA\Desktop\internship\2019_aiq.csv")
df
```

```python
#Select only INDIA
df = df.iloc[56188:74013,:]
df
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Specie"].value_counts()
```

```python
i = df[df.Specie == 'humidity'].index
j = df[df.Specie == 'dew'].index
k = df[df.Specie == 'wind-speed'].index
l = df[df.Specie == 'o3'].index
m = df[df.Specie == 'pm10'].index
n = df[df.Specie == 'pressure'].index
o = df[df.Specie == 'co'].index

p = df[df.Specie == 'wind-gust'].index
q = df[df.Specie == 'wd'].index
r = df[df.Specie == 'precipitation'].index
```

```python
a = [i,j,k,l,m,n,o,p,q,r]
for i in a:
    df = df.drop(i)
    
df["Specie"].value_counts()
```

```python
df
```

```python
#pivoting the table

df = df.pivot_table(['median',"max","min"], ['Date', 'City',"Country"], 'Specie')
df
```

```python
df =  (df.iloc[:, df.columns.get_level_values(0)=='median'])
df
```

```python
df["no2"] = df.iloc[:, df.columns.get_level_values(1)=='no2']
df["so2"] = df.iloc[:, df.columns.get_level_values(1)=='so2']
df["pm25"] = df.iloc[:, df.columns.get_level_values(1)=='pm25']
df["temperature"] = df.iloc[:, df.columns.get_level_values(1)=='temperature']

df
```

```python
#dropping unnecessary columns
df = df.drop(['no2',"pm25","so2","temperature"], axis=1, level=1)
df
```

```python
df.reset_index(inplace = True)
df
```

```python
df.columns = ['_'.join(col) for col in df.columns]
df
```

```python
sns.countplot(df["City_"])
plt.xticks(rotation = 90)
```

```python
#Average So2 in a City_
fig= plt.subplots(figsize=(20, 12))
state_wise_max_so2 = df[['City_','so2_']].dropna().groupby('City_').median().sort_values(by='so2_')
sns.barplot(x='so2_', y=state_wise_max_so2.index, data=state_wise_max_so2)
plt.title("Average So2 in a City before lockdown",fontsize = 30)
plt.xlabel("So2",fontsize = 30)
plt.ylabel("City",fontsize = 30)
plt.show()
```

```python
#Average no2 in a City_
fig= plt.subplots(figsize=(20, 12))
state_wise_max_so2 = df[['City_','no2_']].dropna().groupby('City_').median().sort_values(by='no2_')
sns.barplot(x='no2_', y=state_wise_max_so2.index, data=state_wise_max_so2)
plt.title("Average No2 in a City before lockdown",fontsize = 30)
plt.xlabel("No2",fontsize = 30)
plt.ylabel("City",fontsize = 30)
plt.show()
```

```python
#Average pm25 in a City_
fig= plt.subplots(figsize=(20, 12))
state_wise_max_so2 = df[['City_','pm25_']].dropna().groupby('City_').median().sort_values(by='pm25_')
sns.barplot(x='pm25_', y=state_wise_max_so2.index, data=state_wise_max_so2)
plt.title("Average pm25 in a City before lockdown",fontsize = 30)
plt.xlabel("Pm25",fontsize = 30)
plt.ylabel("City",fontsize = 30)
plt.show()
```

```python
#top 10 highest risk of Respiratory diseases (pm25)

state = df[['City_','pm25_']].groupby(['City_']).median().reset_index()
state_location_max_rspm = state.loc[state.groupby('City_')['pm25_'].idxmax()].sort_values(by='pm25_', ascending=False).head(10)
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(x='pm25_', y= 'City_', data=state_location_max_rspm, palette='coolwarm', axes=ax)
sns.despine(left=True)
ax.set_title("Bars showing average pm25_ values of the cities before lockdown")
plt.tight_layout()
```
